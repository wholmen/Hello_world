Importing data from C++ program to make plots. 

First program has $n=10$

```python
from pylab import *

def readfile(title):
    infile = open(title,'r')
    x = []; v = []; u = []; eps = []; v2 = []
    infile.readline()
    for line in infile:
        splitline = line.split()
        x.append(float(splitline[0])); v.append(float(splitline[1])); u.append(float(splitline[2])); eps.append(float(splitline[3]))
        t1 = float(splitline[4]); t2 = float(splitline[5]); v2.append(float(splitline[6]))
    x = array(x); v = array(v); u = array(u); eps = array(eps); v2 = array(v2)
    infile.close()
    return x,v,u,eps,v2

x,v,u,eps1,v2 = readfile("b_n10.txt")
plot(x,v,color="blue",label="numerical")
plot(x,u,color="green",label="analytical")
plot(x,v2,color="red",label="LU")
legend(loc="upper right")
show()
plot(eps,color="red",label="numerical")
legend(loc="upper right")
show()
```
