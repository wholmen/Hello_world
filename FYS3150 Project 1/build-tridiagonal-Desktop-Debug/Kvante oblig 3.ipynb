```python
from pylab import *

n = 1000
E = 1
g = linspace(0,1,n)

lambda1 = zeros(n)
lambda2 = E/2.*(1+sqrt(1+8*g**2/E**2))
lambda3 = E/2.*(1-sqrt(1+8*g**2/E**2))

plot(g,lambda1, color="blue", label="lambda 1")
plot(g,lambda2, color="red", label="lambda 2")
plot(g,lambda3, color="green", label="lambda 3")
legend(loc='upper left')
show()
```
